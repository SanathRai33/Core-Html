    <h1>Full Stack Web Development Tech Stack</h1>

    <dl>
        <dt><b>HTML (HyperText Markup Language)</b></dt>
        <dd>Defines the structure of web pages using elements like headings, paragraphs, and links.</dd>

        <dt><b>CSS (Cascading Style Sheets)</b></dt>
        <dd>Used for styling web pages, including layout, colors, fonts, and responsiveness.</dd>

        <dt><b>JavaScript</b></dt>
        <dd>A scripting language that adds interactivity and dynamic behavior to web pages.</dd>

        <dt><b>React.js</b></dt>
        <dd>A JavaScript library for building user interfaces, commonly used for single-page applications.</dd>

        <dt><b>Node.js</b></dt>
        <dd>A runtime environment that allows JavaScript to run on the server-side.</dd>

        <dt><b>Express.js</b></dt>
        <dd>A lightweight web framework for Node.js used to build server-side applications and APIs.</dd>

        <dt><b>MongoDB</b></dt>
        <dd>A NoSQL database that stores data in flexible JSON-like documents.</dd>
    </dl>
